```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import polars as pl
```

```python
AK_SHAPs = pl.read_csv("SHAPAnalysis/Data/SHAP/Fastshap/SHAPs/Alaska/AllYearsCombined/AlaskaAllSHAPs.csv")
```

```python
# AK_SHAPs_trim = AK_SHAPs[['peak_swe', 'AbsSHAP_OctMay_prcpSumCDMSum', 'AbsSHAP_OctMay_tmeanCDMSum', 'AbsSHAP_DecFeb_prcpSumCDMSum', 'AbsSHAP_DecFeb_tmeanCDMSum', 'AbsSHAP_elevation', 'AbsSHAP_slope', 'AbsSHAP_aspect', 'AbsSHAP_OctMay_tminMean', 'AbsSHAP_OctMay_tmaxMean', 'AbsSHAP_landcover_triclass']]

AK_SHAPs_trim = AK_SHAPs.select(['peak_swe', 'AbsSHAP_OctMay_prcpSumCDMSum', 'AbsSHAP_OctMay_tmeanCDMSum', 'AbsSHAP_DecFeb_prcpSumCDMSum', 'AbsSHAP_DecFeb_tmeanCDMSum', 'AbsSHAP_elevation', 'AbsSHAP_slope', 'AbsSHAP_aspect', 'AbsSHAP_OctMay_tminMean', 'AbsSHAP_OctMay_tmaxMean', 'AbsSHAP_landcover_triclass'])

```

```python
sil_scores = []
kmax = 25

for k in range(2, kmax+1):
    temp_kmeans = KMeans(n_clusters = k, random_state = 802).fit(AK_SHAPs_trim)
    labels = temp_kmeans.labels_
    sil_scores.append(silhouette_score(AK_SHAPs_trim, labels, metric = "euclidean"))
```
